```python
#Creation of Dictionary
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
print(thisdict)
```

```python
#Retriving values from dictionary
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
print(thisdict["brand"])
```

```python
#Dictionary takes latest value given for key.It will not allow duplicates
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964,
    "year": 2020
}
print(thisdict)

# Length of key-value pairs
len(thisdict)
```

```python
# Changing/Adding key:value in dictionary
car = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
car["color"] = "red"
print(car)
x = car.values()
print(x)
print(type(car))
```
